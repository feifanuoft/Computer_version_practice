```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import sklearn

print(cv2.__version__)
print(np.__version__)
print(sklearn.__version__)
```

```python
print("Hello World!")
img =  cv2.imread("lena.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title("Lena Color")
```

```python
plt.imshow(gray, cmap='gray')
plt.title("Lena Grayscale")
```

```python
print(img.shape)
print(img.size)
print(img.dtype)
```

```python
i = 0
j = 0
k = 0
imgh = np.zeros((512,512,3),np.uint8)
for k in range(0,3):
    for j in range(0,512):
        for i in range(0,512):
            imgh[i,j,k] = img[i,512-1-j,k]
plt.imshow(cv2.cvtColor(imgh,cv2.COLOR_BGR2RGB))
plt.title("Lena color horizontal flip")
```

```python
i = 0
j = 0
k = 0
imgv = np.zeros((512,512,3),np.uint8)
for k in range(0,3):
    for j in range(0,512):
        for i in range(0,512):
            imgv[i,j,k] = img[512-1-i,j,k]
plt.imshow(cv2.cvtColor(imgv,cv2.COLOR_BGR2RGB))
plt.title("Lena color vertical flip")
```

```python
i = 0
j = 0
k = 0
imghg = np.zeros((512,512,3),np.uint8)
for k in range(0,3):
    for j in range(0,512):
        for i in range(0,512):
            imghg[i,j,k] = img[i,512-1-j,k]
imghg= cv2.cvtColor(imghg,cv2.COLOR_BGR2GRAY)
plt.imshow(imghg, cmap = 'gray')
plt.title("Lena horizontal flip")
```

```python
i = 0
j = 0
k = 0
imgvg = np.zeros((512,512,3),np.uint8)
for k in range(0,3):
    for j in range(0,512):
        for i in range(0,512):
            imgvg[i,j,k] = img[512-1-i,j,k]
imgvg= cv2.cvtColor(imgvg,cv2.COLOR_BGR2GRAY)
plt.imshow(imgvg, cmap = 'gray')
plt.title("Lena vertical flip(gray)")
```

```python
imgneg = np.zeros((512,512,3),np.uint8)
for i in range (0,512):
    for j in range(0,512):
        (B,G,R) = img[i,j]
        B = 255-B
        G = 255-G
        R = 255-R
        imgneg[i,j] = (B,G,R)
plt.imshow(cv2.cvtColor(imgneg,cv2.COLOR_BGR2RGB))
plt.title("Lena Color negated")
```

```python
imgneggray = np.zeros((512,512,3),np.uint8)
for i in range (0,512):
    for j in range(0,512):
        (B,G,R) = img[i,j]
        B = 255-B
        G = 255-G
        R = 255-R
        imgneggray[i,j] = (B,G,R)
imgneggray = cv2.cvtColor(imgneggray,cv2.COLOR_BGR2GRAY)
plt.imshow(imgneggray, cmap = 'gray')
plt.title("Lena Color negated")
```

```python
blueimf = np.zeros((512,512,3),np.uint8)
for i in range (0,512):
    for j in range(0,512):
        (B,G,R) = img[i,j]
        blueimf[i,j] = (B,0,0)
        
plt.imshow(cv2.cvtColor(blueimf,cv2.COLOR_BGR2RGB))
plt.title("Lena blue image")
```

```python
greenimf = np.zeros((512,512,3),np.uint8)
for i in range (0,512):
    for j in range(0,512):
        (B,G,R) = img[i,j]
        greenimf[i,j] = (0,G,0)
        
plt.imshow(cv2.cvtColor(greenimf,cv2.COLOR_BGR2RGB))
plt.title("Lena green image")
```

```python
redimf = np.zeros((512,512,3),np.uint8)
for i in range (0,512):
    for j in range(0,512):
        (B,G,R) = img[i,j]
        redimf[i,j] = (0,0,R)
        
plt.imshow(cv2.cvtColor(redimf,cv2.COLOR_BGR2RGB))
plt.title("Lena red image")
```

```python
mergeimf = np.zeros((512,512,3),np.uint8)
for i in range (0,512):
    for j in range(0,512):
        (B1,G1,R1) = blueimf[i,j]
        (B2,G2,R2) = greenimf[i,j]
        (B3,G3,R3) = redimf[i,j]
        B = B1+B2+B3
        G = G1+G2+G3
        R = R1+R2+R3
        mergeimf[i,j] = (B,G,R)
plt.imshow(cv2.cvtColor(mergeimf,cv2.COLOR_BGR2RGB))
plt.title("Merged Lena image ")
        
        
```
