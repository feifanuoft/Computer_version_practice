Import

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
my_SIFT_instance = cv2.SIFT_create()
```

```python
cer_l = cv2.imread('cereal_l.jpg')
cer_l = cv2.cvtColor(cer_l,cv2.COLOR_BGR2GRAY)
cer_r = cv2.imread('cereal_r.jpg')
cer_r = cv2.cvtColor(cer_r,cv2.COLOR_BGR2GRAY)
```

```python
def loweratiomatch(img1,img2):
    kp1, des1 = my_SIFT_instance.detectAndCompute(img1,None)
    kp2, des2 = my_SIFT_instance.detectAndCompute(img2,None)
    fbm = cv2.FlannBasedMatcher()
    matches = fbm.knnMatch(des1,des2, k=2)
    good = []
    goodlist=[]
    for m,n in matches:
        if(m.distance < 0.75*n.distance):
            good.append([m])
            goodlist.append(m)
    return kp1,kp2,des1,des2,good,goodlist
```

```python
kp1,kp2,des1,des2,goodcer,goodcerlist= loweratiomatch(cer_l,cer_r)
plt.figure(figsize=(15,5))
cerconnect = cv2.drawMatchesKnn(cer_l,kp1,cer_r,kp2,goodcer,None,flags=2)
plt.title("1.1 Matching")
plt.imshow(cerconnect)
```

```python
cer = cv2.imread('cereal.jpg')
cergray = cv2.cvtColor(cer,cv2.COLOR_BGR2GRAY)
center  = [cer.shape[1]/2,cer.shape[0]/2]
print(center)
rotation = 15
scaling = 0.6
rotmatrix  = cv2.getRotationMatrix2D(center, rotation, scaling)
img_rotated= cv2.warpAffine(cergray, rotmatrix, (cer.shape[1], cer.shape[0]))
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.title("Reference image")
plt.imshow(cergray,cmap = 'gray')
plt.subplot(1,2,2)
plt.title("Target image")
plt.imshow(img_rotated,cmap = 'gray')
```

```python
ref = cer.copy()
kp1,kp2,des1,des2,goodcerrot,goodcerrotlist = loweratiomatch(cergray,img_rotated)
plt.figure(figsize=(15,5))
cerconnect = cv2.drawMatchesKnn(cergray,kp1,img_rotated,kp2,goodcerrot,None,flags=2)
plt.title("Matching")
plt.imshow(cerconnect)    
```

```python

ref_pts = np.float32([kp1[m.queryIdx].pt for m in goodcerrotlist]).reshape(-1, 1, 2)
img_pts = np.float32([kp2[m.trainIdx].pt for m in goodcerrotlist]).reshape(-1, 1, 2)
transformation_matrix, rigid_mask = cv2.estimateAffinePartial2D(ref_pts, img_pts)
image_t = cv2.warpAffine(ref, transformation_matrix, (cer.shape[1], cer.shape[0]))
image_t= cv2.cvtColor(image_t,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,5))
plt.imshow(image_t)
```

```python
certl=cv2.imread("cereal_tl.jpg")
certl = cv2.cvtColor(certl,cv2.COLOR_BGR2GRAY)
certr=cv2.imread("cereal_tr.jpg")
certr = cv2.cvtColor(certr,cv2.COLOR_BGR2GRAY)
def perspectivetrans(img1, img2):
    img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
    kp1,kp2,des1,des2,goodcer,goodcerlist= loweratiomatch(img1,img2)
    ref_pts = np.float32([kp1[m.queryIdx].pt for m in goodcerlist]).reshape(-1, 1, 2)
    img_pts = np.float32([kp2[m.trainIdx].pt for m in goodcerlist]).reshape(-1, 1, 2)
    transformation_matrix, rigid_mask = cv2.findHomography(ref_pts, img_pts, method = cv2.RANSAC)
    image_t = cv2.warpPerspective(ref, transformation_matrix, (img2.shape[1], img2.shape[0]),borderMode = cv2.BORDER_CONSTANT)
    image_t= cv2.cvtColor(image_t,cv2.COLOR_BGR2RGB)
    temp = cv2.cvtColor(img2,cv2.COLOR_GRAY2BGR)
    for i in range(img2.shape[1]):
        for j in range(img2.shape[0]):
            if image_t[j][i].all()==0:
                image_t[j][i]=temp[j][i]
    return image_t
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.title("Reference image")
plt.imshow(cergray,cmap = 'gray')
plt.subplot(1,3,2)
plt.title("Target image")
plt.imshow(cer_l,cmap = 'gray')
plt.subplot(1,3,3)
cer_ltrans = perspectivetrans(ref, cer_l)
plt.title("Overlay from recovered homography")
plt.imshow(cer_ltrans)
```
