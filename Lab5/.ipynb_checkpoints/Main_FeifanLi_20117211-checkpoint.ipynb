## Import

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
my_SIFT_instance = cv2.SIFT_create()


bl = cv2.imread("bike_l.png")
bl = cv2.cvtColor(bl,cv2.COLOR_BGR2RGB)
br = cv2.imread("bike_r.png")
br = cv2.cvtColor(br,cv2.COLOR_BGR2RGB)
bimgsize = (bl.shape[1], bl.shape[0])
cl = cv2.imread("coaster_left.jpg")
cl = cv2.cvtColor(cl,cv2.COLOR_BGR2RGB)
cr = cv2.imread("coaster_right.jpg")
cr = cv2.cvtColor(cr,cv2.COLOR_BGR2RGB)
tl = cv2.imread("tape_l.jpg")
tl = cv2.cvtColor(tl,cv2.COLOR_BGR2RGB)
tr = cv2.imread("tape_r.jpg")
tr = cv2.cvtColor(tr,cv2.COLOR_BGR2RGB)
```

```python
def loweratiomatch(img1,img2):
    kp1, des1 = my_SIFT_instance.detectAndCompute(img1,None)
    kp2, des2 = my_SIFT_instance.detectAndCompute(img2,None)
    fbm = cv2.FlannBasedMatcher()
    matches = fbm.knnMatch(des1,des2, k=2)
    good = []
    goodlist=[]
    pts1 = []
    pts2 = []
    for m,n in matches:
        if(m.distance < 0.8*n.distance):
            good.append([m])
            goodlist.append(m)
            pts1.append(kp1[m.queryIdx].pt)  
            pts2.append(kp2[m.trainIdx].pt)

    return kp1,kp2,des1,des2,good,goodlist,pts1,pts2

def FundamentalMatrixCal(img1, img2):
    point1 = []
    point2 = []
    kp1,kp2,des1,des2,good,goodlist,point1,point2 = loweratiomatch(img1,img2)
    point1 = np.int32(point1)
    point2 = np.int32(point2)
    F, mask = cv2.findFundamentalMat(point1, point2, cv2.FM_LMEDS)
    
    return F, point1.flatten(),point2.flatten()
```

```python
def ImgRectification(img1, img2):
    
    F, point1, point2 = FundamentalMatrixCal(img1, img2)
    _, h1, h2 = cv2.stereoRectifyUncalibrated(point1, point2,F,(img1.shape[0], img1.shape[1]), threshold = 10)
    leftimg = cv2.warpPerspective(cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY),h1,(img1.shape[1], img1.shape[0]), borderValue = (0,0,0))
    rightimg = cv2.warpPerspective(cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY),h2,(img1.shape[1], img1.shape[0]), borderValue = (0,0,0))
    return leftimg,rightimg
```

## 1.2

```python
lout1, rout1 = ImgRectification(tl,tr)
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
plt.imshow(lout1,cmap="gray")
plt.title("Left Image")
plt.subplot(1,2,2)
plt.imshow(rout1,cmap="gray")
plt.title("Right Image")
```

```python
lout2, rout2 = ImgRectification(cl,cr)
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
plt.imshow(lout2,cmap="gray")
plt.title("Left Image")
plt.subplot(1,2,2)
plt.imshow(rout2,cmap="gray")
plt.title("Right Image")
```

```python
lout3, rout3 = ImgRectification(bl,br)
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
plt.imshow(lout3,cmap="gray")
plt.title("Left Image")
plt.subplot(1,2,2)
plt.imshow(rout3,cmap="gray")
plt.title("Right Image")
```

```python
def BlockMatchingDisparity(img1,img2, numDispar, bSize):
    my_stereo_instance = cv2.StereoBM_create(numDisparities = numDispar, blockSize = bSize)
    outImg = my_stereo_instance.compute(img1, img2)
    return outImg
outimg1 = BlockMatchingDisparity(lout1,rout1, 32,5)
outimg2 = BlockMatchingDisparity(lout2,rout2, 32,5)
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
plt.imshow(outimg1,cmap="gray")
plt.title("Left Image")
plt.subplot(1,2,2)
plt.imshow(outimg2,cmap="gray")
plt.title("Right Image")
```
