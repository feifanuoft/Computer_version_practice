## Step1: find Best pair

```python
def loweratiomatch(img1,img2):
    kp1, des1 = my_SIFT_instance.detectAndCompute(img1,None)
    kp2, des2 = my_SIFT_instance.detectAndCompute(img2,None)
    fbm = cv2.FlannBasedMatcher()
    matches = fbm.knnMatch(des1,des2, k=2)
    good = []
    goodlist=[]
    for m,n in matches:
        if(m.distance < 0.75*n.distance):
            good.append([m])
            goodlist.append(m)
    goodcount  =  len(goodlist)
    loweresult = [kp1,kp2,des1,des2,good,goodlist,goodcount]
    return loweresult


def get_best(imgs):
    threshold = 0
    combine_list=[]
    for i in range (len(imgs)-1, 0, -1):
        cur_list =[]
        for j in range(i-1, -1,-1 ):  # Construct a combination matrix 
            cur_list.append([i,j])
        combine_list.append(cur_list)
    result =[]                         # store the result (pairs)
    
    for item in combine_list:
        for item_1 in item:
            if(loweratiomatch(imgs[item_1[0]],imgs[item_1[1]])[6] > threshold ):
                item_1.append(loweratiomatch(imgs[item_1[0]],imgs[item_1[1]])[6]) 
        item.sort(key=lambda x:x[2],reverse=True)
        result.append(item[0])
    return result
    print(result)
    
start = time.perf_counter()
result = get_best(imgs)
print("finish")
print(result)
for i in range(len(result)):
    result[i].pop(2)
    result[i].reverse()
result.reverse()

elapsed = (time.perf_counter() - start)
print("Time used:",elapsed)
print(result)
```

## Step2: Transform function

```python

def transform(img1,img2):
    result = loweratiomatch(img1,img2)
    goodmatch= result[4]
    ref_pts = np.float32([result[0][m.queryIdx].pt for m in result[5]]).reshape(-1, 1, 2)
    img_pts = np.float32([result[1][m.trainIdx].pt for m in result[5]]).reshape(-1, 1, 2)
    transformation_matrix, rigid_mask = cv2.findHomography(ref_pts, img_pts, method = cv2.RANSAC)
    
    return (goodmatch,transformation_matrix, rigid_mask)
```

## Step3: Merge

```python
def blendingMask(height, width, barrier, smoothing_window, left=True):
    assert barrier < width
    mask = np.zeros((height, width))

    offset = int(smoothing_window / 2)

    if left:
        mask[:, barrier - offset : barrier + offset + 1] = np.tile(
            np.linspace(1, 0, 2 * offset + 1).T, (height, 1)
        )
        mask[:, : barrier - offset] = 1
    else:
        mask[:, barrier - offset : barrier + offset + 1] = np.tile(
            np.linspace(0, 1, 2 * offset + 1).T, (height, 1)
        )
        mask[:, barrier + offset :] = 1

    return cv2.merge([mask, mask, mask])


def Blending(dst_img_rz, src_img_warped, width_dst, side):

    h, w, _ = dst_img_rz.shape
    smoothing_window = int(width_dst / 8)
    barrier = width_dst - int(smoothing_window / 2)
    mask1 = blendingMask(
        h, w, barrier, smoothing_window=smoothing_window, left=True
    )
    mask2 = blendingMask(
        h, w, barrier, smoothing_window=smoothing_window, left=False
    )


    if side == "left":
        dst_img_rz = cv2.flip(dst_img_rz, 1)
        src_img_warped = cv2.flip(src_img_warped, 1)
        dst_img_rz = dst_img_rz * mask1
        src_img_warped = src_img_warped * mask2
        pano = src_img_warped + dst_img_rz
        pano = cv2.flip(pano, 1)
    else:
        dst_img_rz = dst_img_rz * mask1
        src_img_warped = src_img_warped * mask2
        pano = src_img_warped + dst_img_rz


    return pano

def crop(panorama, h_dst, conners):

    # find max min of x,y coordinate
    [xmin, ymin] = np.int32(conners.min(axis=0).ravel() - 0.5)
    t = [-xmin, -ymin]
    conners = conners.astype(int)
    
    ## Judge the imgs are merged to left or right sides
    if conners[0][0][0] < 0:
        n = abs(-conners[1][0][0] + conners[0][0][0])
        panorama = panorama[t[1] : h_dst + t[1], n:, :]
    else:
        if conners[2][0][0] < conners[3][0][0]:
            panorama = panorama[t[1] : h_dst + t[1], 0 : conners[2][0][0], :]
        else:
            panorama = panorama[t[1] : h_dst + t[1], 0 : conners[3][0][0], :]
    return panorama
    

def Merge(img1,img2):
    _,H, _ = transform(img1, img2)

    # get row and column of two images
    row_src, col_src = img1.shape[:2]
    row_dst, col_dst = img2.shape[:2]

    # extract conners of two images: top-left, bottom-left, bottom-right, top-right
    pts1 = np.float32([[0, 0], [0, row_src], [col_src, row_src], [col_src, 0]]).reshape(-1, 1, 2)   
    pts2 = np.float32([[0, 0], [0, row_dst], [col_dst, row_dst], [col_dst, 0]]).reshape(-1, 1, 2)

    # aply homography to conners of src_img
    pts1_ = cv2.perspectiveTransform(pts1, H)
    pts = np.concatenate((pts1_, pts2), axis=0)

    # find max min of x,y coordinate
    [xmin, ymin] = np.int64(pts.min(axis=0).ravel() - 0.5)
    [_, ymax] = np.int64(pts.max(axis=0).ravel() + 0.5)
    t = [-xmin, -ymin]

    if pts[0][0][0] < 0:
        side = "left"
        col_pano = col_dst + t[0]
    else:
        col_pano = int(pts1_[3][0][0])
        side = "right"
    row_pano = ymax - ymin

    Ht = np.array([[1, 0, t[0]], [0, 1, t[1]], [0, 0, 1]])
    src_img_warped = cv2.warpPerspective(
        img1, Ht.dot(H), (col_pano, row_pano)
    )
    dst_img_rz = np.zeros((row_pano, col_pano, 3))
    if side == "left":
        dst_img_rz[t[1] : row_src + t[1], t[0] : col_dst + t[0]] = img2
    else:
        dst_img_rz[t[1] : row_src + t[1], :col_dst] = img2

    pano= Blending(
        dst_img_rz, src_img_warped, col_dst, side
    )

    pano = crop(pano, row_dst, pts)

    return pano
```

```python



def Multimerge(imgs):

    n = int(len(imgs) / 2 + 0.5)
    left = imgs[:n]
    right = imgs[n - 1 :]
    right.reverse()
    while len(left) > 1:
        dst_img = left.pop()
        src_img = left.pop()
        left_pano = Merge(src_img, dst_img)
        left_pano = left_pano.astype("uint8")
        left.append(left_pano)

    while len(right) > 1:
        dst_img = right.pop()
        src_img = right.pop()
        right_pano = Merge(src_img, dst_img)
        right_pano = right_pano.astype("uint8")
        right.append(right_pano)

    if right_pano.shape[1] >= left_pano.shape[1]:
        fullpano = Merge(left_pano, right_pano)
    else:
        fullpano= Merge(right_pano, left_pano)
    return fullpano

def convertimg(img):
    img = np.array(img,dtype=float)/float(255)
    return img

start = time.perf_counter()
mergetest=Merge(sortimg[0],sortimg[1])
plt.figure(figsize=(15,10))
plt.imshow(convertimg(mergetest))
elapsed = (time.perf_counter() - start)
print("Time used:",elapsed)
```

## Stjames

```python
my_SIFT_instance = cv2.SIFT_create()
pBarLength = 70

imgsjames = [cv2.imread(file) for file in glob.glob("StJames\\*.jpg")]
print(len(imgsjames))
imgjames = []
imggrayjames = []
for i in range(len(imgsjames)):
    imgsjames[i] = cv2.cvtColor(imgsjames[i],cv2.COLOR_BGR2RGB)
    imgjames.append(imgsjames[i])
    imgsjames[i] = cv2.resize(imgsjames[i], (int(imgsjames[i].shape[1] / 2), int(imgsjames[i].shape[0] / 2)))
    imggrayjames.append(cv2.cvtColor(imgjames[i],cv2.COLOR_BGR2GRAY))
```

```python
import statistics
def goodmatch_for_St(imgs):
            
    sift = cv2.xfeatures2d.SIFT_create()
    fbm = cv2.FlannBasedMatcher()
    
    kp_all = []
    des_all = []
    
    for m in range(len(imgs)):
        kp, ds = sift.detectAndCompute(imgs[m], None) 
        kp_all.append(kp)
        des_all.append(ds)


    def get_good(kp1,kp2,ds1,ds2,fbm):
        matches = fbm.knnMatch(ds1,ds2,k=2)
        # Apply ratio test
        good = []
        for m, n in matches:
            if(m.distance < 0.75*n.distance):
                good.append(m)

        return good, len(matches), len(good)
    
    goods = []
    good_matchcounts=[]
    for i in range(len(kp_all)):
        goods.append([])
        for k in range(i+1,len(kp_all)):
            good, t_match, g_match = get_good(kp_all[i],kp_all[k],des_all[i],des_all[k],fbm)
            goods[i].append([good, t_match, g_match, i, k])
            good_matchcounts.append(g_match)
            
    mean=statistics.mean(good_matchcounts)  # mean value is threshold 


    goodpairs=[]
    for i in range(len(goods)):
        for j in range(len(goods[i])):
            if goods[i][j][2] > mean:
                goodpairs.append(goods[i][j])
    imgpts_low = [np.int32([kp_all[i[3]] [m.queryIdx].pt for m in i[0]]).reshape(-1, 1, 2) for i in goodpairs]
    imgpts_high = [np.int32([kp_all[i[4]] [m.trainIdx].pt for m in i[0]]).reshape(-1, 1, 2) for i in goodpairs]
    
    F=[]
    mask = []

    for i in range(len(imgpts_low)):
        transformation_matrix,rigid_mask = cv2.findHomography(imgpts_low[i],imgpts_high[i],method = cv2.RANSAC)
        F.append(transformation_matrix)
        mask.append(rigid_mask)

    #how many times are each the image are used
    temp =[]
    temp = [temp + [goodpairs[i][3] , goodpairs[i][4]] for i in range(len(goodpairs))]
    
    return temp, goodpairs,F,mask
    
start = time.perf_counter()
temp, goodpairs,F,mask = goodmatch_for_St(imgsjames)
elapsed = (time.perf_counter() - start)
print("Time used:",elapsed)
print(temp)
```
