```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import glob
import os
my_SIFT_instance = cv2.SIFT_create()
```

```python
imgs = []
img_dir = "C:\\Users\\53061\\ELEC474\\FinalProject\\office2" # Enter Directory of all images  
data_path = os.path.join(img_dir,'*jpg') 
files = glob.glob(data_path) 
for f1 in files: 
    img = cv2.imread(f1) 
    imgs.append(img) 
```

```python
def loweratiomatch(img1,img2):
    kp1, des1 = my_SIFT_instance.detectAndCompute(img1,None)
    kp2, des2 = my_SIFT_instance.detectAndCompute(img2,None)
    fbm = cv2.FlannBasedMatcher()
    matches = fbm.knnMatch(des1,des2, k=2)
    good = []
    goodlist=[]
    pts1 = []
    pts2 = []
    for m,n in matches:
        if(m.distance < 0.85*n.distance):
            good.append([m])
            goodlist.append(m)
            pts1.append(kp1[m.queryIdx].pt)  
            pts2.append(kp2[m.trainIdx].pt)

    return kp1,kp2,des1,des2,good,goodlist,pts1,pts2

kp1,kp2,des1,des2,good,goodlist,pts1,pts2 = loweratiomatch(imgs[0],imgs[1])

plt.figure(figsize=(15,15))
cerconnect = cv2.drawMatchesKnn(imgs[0],kp1,imgs[1],kp2,good,None,flags=2)
plt.title("1.1 Matching")
plt.imshow(cerconnect)
```
